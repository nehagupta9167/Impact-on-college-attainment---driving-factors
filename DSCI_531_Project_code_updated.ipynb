```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
sns.lineplot(data = parent_income_percentile, x = "par_bin", y = "college", ax = ax)
ax.set_xlabel("Income Percentile of parent family")
ax.set_ylabel("Sharing of children attending college")
ax.set_title("Figure 1")
plt.show()
```

```python
fig, ax = plt.subplots()
sns.lineplot(data = parent_income_percentile, x = "par_bin", y = "par_fam_inc", ax = ax)
ax.set_xlabel("Income Percentile of parent family")
ax.set_ylabel("Parental Family Income")
ax.set_title("Figure 2")
plt.show()
```

```python
fig, ax = plt.subplots()
sns.lineplot(data = parent_income_percentile, x = "par_fam_inc", y = "college", ax = ax)
ax.set_xlabel("Parental Family Income")
ax.set_ylabel("Kids attending college during age 18-21")
ax.set_title("Figure 3")
plt.show()
```

> In 2019, the US median family income was estimated at $86,011 by the US Census Bureau.

> we see that the 90\% mark is reached at a household income of almost $200 000 

```python
fig, ax = plt.subplots()
sns.lineplot(data = parent_income_percentile, x = "par_fam_inc", y = "kid_fam_inc", ax = ax)
ax.set_xlabel("Parental Family Income")
ax.set_ylabel("Average kid family income")
ax.set_title("Figure 4")
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
sns.lineplot(data = parent_income_percentile, x = "par_fam_inc", y = "coll_qual_20_rank", ax = ax)
ax.set_xlabel("Parental Family Income")
ax.set_ylabel("College quality rank")
ax.set_title("Figure 5")
plt.grid()
plt.show()

"""
Count of people in study with income greater than $200,000 = 493387. Only 5% of the total considered.
Only kids born into family above this income bracket has chances of attending college which were having quality rank greater than .75

Count of remaining people = 9374349
"""
```
